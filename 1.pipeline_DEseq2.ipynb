```r
library( "DESeq2" )
#library(ggplot2)
```

```r
countData <- read.csv('sc_counts.csv', header = TRUE, sep = ",")
metaData <- read.csv('sc_meta.csv', header = TRUE, sep = ",")
dds <- DESeqDataSetFromMatrix(countData=countData, 
                              colData=metaData, 
                              design=~type, tidy = TRUE)
dds.sizefactor <- estimateSizeFactors(dds)
## 如果想要知道具体的sizefactor是多少，可以使用sizeFactors()函数
sizeFactors(dds.sizefactor )
```

```r
dds <- estimateSizeFactors(dds)
counts(dds,normalized=TRUE)
dds
```

```r
write.csv(counts(dds,normalized=TRUE),file='NEGF_normdata.csv')
```

```r
countData <- read.csv('NEGF_counts.csv', header = TRUE, sep = ",")
metaData <- read.csv('NEGF_meta.csv', header = TRUE, sep = ",")
dds <- DESeqDataSetFromMatrix(countData=countData, 
                              colData=metaData, 
                              design=~type, tidy = TRUE)
dds <- DESeq(dds)
res <- results(dds)

res <- res[order(res$padj),]
write.csv(res,file='NEGF_result11.csv')
```
